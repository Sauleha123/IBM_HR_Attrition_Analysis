### Attrition Distribution

```python
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print("\n Attrition Rate (%): \n", attrition_rate)

plt.figure(figsize=(6,4))
ax = sns.countplot(data=df, x='Attrition', palette='Set2')

total = len(df)
for p in ax.patches:
    count = p.get_height()
    percentage = count / total * 100
    ax.annotate(f'{percentage: .1f}%',
                (p.get_x() + p.get_width() / 2., count),
                ha='center', va='bottom', fontsize = 12, color = 'black')
    
plt.title('Overall Attrition Percentage')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.show()
```

#### 
- Around 16.1% of employees have left the company, while 83.9% stayed.
- Indicates a relatively moderate attrition rate but still worth investigating causes.

## Handling Outliers
##### Since we cannot remove the outliers completely because it might impact the insights, so we will just plot the graphs representing them with and without outliers.

```python
df['MonthlyIncome'] = pd.to_numeric(df['MonthlyIncome'], errors='coerce')

Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['Outlier_Flag'] = ((df['MonthlyIncome'] < lower_bound) | (df['MonthlyIncome'] > upper_bound)).astype(int)
print(f"Number of Outlier: {df['Outlier_Flag'].sum()}")

df_no_outliers = df[df['Outlier_Flag'] == 0]

plt.figure(figsize=(14,6))

# Box Plot With Outliers

plt.subplot(1,2,1)
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette='pastel')
plt.title('Monthly Income vs Attrition (With Outliers)')

# Box Plot Without Outliers

plt.subplot(1,2,2)
sns.boxplot(data=df_no_outliers, x='Attrition', y='MonthlyIncome', palette='pastel', showfliers=False)
plt.title('Monthly Income vs Attrition (Without Outliers)')

plt.tight_layout()
plt.show()
```

#### 
- Outliers in Monthly Income are primarily on the higher end, representing a small group of higher paid employees.
- Removing outliers reduces the spread but does not significantly affect the median income for either attrition group.
- High-income outliers are more prevalent among employees who stayed, suggesting better compensation may contribute to retention.

## Key Insights:
The analysis of IBM HR Employee Attrition data reveals several key factors influencing employee turnover:

- Income & Compensation:
Lower monthly income is strongly associated with higher attrition. High-income employees are more likely to stay, and income outliers (very high salaries) are mostly in the non-attrition group.

- Tenure & Experience:
Employees with fewer years at the company, especially under 3 years, have significantly higher attrition rates. Long-term employes (10+ years) rarely leave.

- Age Factor:
Younger employees (20-30 years) show the highest attrition, suggesting career exploration, lack of engagement, or better opportunities elsewhere. Attrition decreases steadily with age.

- Job Roles & Departments:
Certain roles particularly Sales Executive, Research Scientist, and Laboratory Technician face higher attrition. Human Resources shows the lowest turnover.

- Overtime & Workload:
Overtime workers have substantially higher attrition rates, indicating that workload and work-life balance play an important role in retention.

- Satisfaction Levels:
Lower job satisfaction and environment scores (1-2) correlate with higher attrition, highlighting the importance of employee engagement and workplace environment.

- Gender Distribution:
While males slightly outnumber females in the workforce, attrition appears more influenced by role,pay, and satisfaction than by gender.


### Overall, the findings suggest that IBM can reduce attrition by:
- Improving pay structures for lower-income employees.
- Addressing workload and overtime concerns.
- Focusing on engagement and satisfaction in high-turnover job roles.
- Implementing targeted retention programs for younger and early-tenure employees.
